Using word cloud generator from https://github.com/amueller/word_cloud (documentation at https://amueller.github.io/word_cloud/index.html).

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Read the whole text.
text = open('facebook-mining-words.txt').read()
```

```python
stopwords = set(STOPWORDS)
stopwords.add("simple_smile")
stopwords.add("face_with_tears_of_joy")
stopwords.add("smiling_face_with_smiling_eyes")
stopwords.add("yes")
stopwords.add("one")
stopwords.add("smiley")
stopwords.add("beaming_face_with_smiling_eyes")
stopwords.add("grinning_face_with_smiling_eyes")
stopwords.add("will")
stopwords.add("pensive")

wc = WordCloud(background_color = "white", max_words = 2000,
               stopwords = stopwords, width = 500, height = 500,
               colormap = plt.cm.gray)

# Generate word cloud.
wc.generate(text)

# Store to file.
wc.to_file("wordcloud2.png")

# Show plt.
plt.imshow(wc, interpolation = 'bilinear')
plt.axis("off")
plt.figure()
plt.imshow(wc, cmap = plt.cm.gray, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
